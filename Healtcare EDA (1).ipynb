```python
# DATA PREPARATION

#in data preparation we drop irrelevant columns and rows
#identify duplicated columns
#if needed renaming the columns
#look for missing values
#handle any inconsistent entries
#if needed creating new features

 

```

```python
#Dropping non informative id columns
df.drop(columns=['encounter_id', 'patient_nbr'], inplace=True, errors='ignore')
#these columns have the same value
df.drop(columns=['citoglipton', 'examide'], inplace=True, errors='ignore')
```

```python
# Replace '?' with NaN
df.replace('?', pd.NA, inplace=True)

# Drop rows with unknown/invalid gender
df = df[df['gender'] != 'Unknown/Invalid']
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Remove duplicates (if any)
df = df.drop_duplicates()
```

```python
# time in hospital 
plt.figure(figsize=(8, 6))
sns.histplot(df['time_in_hospital'], kde=True) #Histogram with KDE
plt.title('Distribution of Time Spent in Hospital')
plt.xlabel('Days in Hospital')
plt.ylabel('Frequency')
plt.show()

# number of lab procedures
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['num_lab_procedures'])
plt.title('Box Plot of Number of Lab Procedures')
plt.ylabel('Number of Lab Procedures')
plt.show()
```

```python
#seeing how age grops are divided
sns.histplot(df['age'], bins=10, kde=True, color='cornflowerblue')
plt.title('Age Distribution of Patients')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#seeing which gender has the highest patients

sns.countplot(data=df, x='age', hue='gender')
plt.title('Gender Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

```python
#seeing  how does the number of lab procedures vary across 
#different age groups
#violin plot is similar to boxplot but shows density type of graph

plt.figure(figsize=(10, 6))
sns.violinplot(x=df['age'], y=df['num_lab_procedures'])
plt.title('Number of Lab Procedures by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Lab Procedures')
plt.xticks(rotation=45)
plt.show()
```

```python
#seeing the same grph above in boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['age'], y=df['num_lab_procedures'])
plt.title('Number of Lab Procedures by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Lab Procedures')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot of time and no of medications
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['time_in_hospital'], y=df['num_medications'])
plt.title('Time in Hospital vs. Number of Medications')
plt.xlabel('Time in Hospital')
plt.ylabel('Number of Medications')
plt.show()
```

```python
# time in hospital by gender
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', hue='time_in_hospital',data=df)
plt.title('Time in Hospital by Gender')
plt.xlabel('Gender')
plt.ylabel('Time in Hospital')
plt.show()
```

```python
#the same above plot in boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['gender'], y=df['time_in_hospital'])
plt.title('Time in Hospital by Gender')
plt.xlabel('Gender')
plt.ylabel('Time in Hospital')
plt.show()
```

```python
# plot between readmitted and gender
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', hue='readmitted', data=df)
plt.title('Readmission Status by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Heatmap of age group and gender
#this shows which age groups are more prone 
age_gender = pd.crosstab(df['age'], df['gender'])

plt.figure(figsize=(10,8))
sns.heatmap(age_gender, cmap='Blues', annot=True, fmt='d')
plt.title('Age Group vs Gender')
plt.xlabel('Gender')
plt.ylabel('Age Group')
plt.show()
```

```python
# We can analyse the factors affecting readmission rates

#first we need to analyse the readmitted column
# 1. readmision rate
plt.figure(figsize=(6, 4))
sns.countplot(x='readmitted', data=df)
plt.title('Distribution of Readmission Status')
plt.xlabel('Readmitted')
plt.ylabel('Count')
plt.show()
```

```python
# 2. readmission and time graph
plt.figure(figsize=(8, 6))
sns.boxplot(x='readmitted', y='time_in_hospital', data=df)
plt.title('Readmission Status vs. Time in Hospital')
plt.xlabel('Readmitted')
plt.ylabel('Time in Hospital')
plt.show()
```

```python
# 3. readmission and no of medications
plt.figure(figsize=(8, 6))
sns.boxplot(x='readmitted', y='num_medications', data=df)
plt.title('Readmission Status vs. Number of Medications')
plt.xlabel('Readmitted')
plt.ylabel('Number of Medications')
plt.show()
```

```python
# 4. Readmission and age
plt.figure(figsize=(10, 6))
sns.countplot(x='age', hue='readmitted', data=df)
plt.title('Readmission Status by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# 5. Readmission and no of diagnoses
plt.figure(figsize=(8, 6))
sns.boxplot(x='readmitted', y='number_diagnoses', data=df)
plt.title('Readmission Status vs. Number of Diagnoses')
plt.xlabel('Readmitted')
plt.ylabel('Number of Diagnoses')
plt.show()
```

```python
# 6. readmission vs dischatrsge
plt.figure(figsize=(12, 6))
sns.countplot(x='discharge_disposition_id', hue='readmitted', data=df)
plt.title('Readmission Status by Discharge Disposition')
plt.xlabel('Discharge Disposition ID')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
# Calculate readmission rates for different groups
readmission_counts = df['readmitted'].value_counts()
total_patients = len(df)

readmission_rate_overall = (readmission_counts['<30'] + readmission_counts['>30']) / total_patients
print(f"Overall Readmission Rate: {readmission_rate_overall:.2f}")

readmission_by_gender = df.groupby('gender')['readmitted'].value_counts(normalize=True).unstack()
print("\nReadmission Rates by Gender:\n", readmission_by_gender)

readmission_by_age = df.groupby('age')['readmitted'].value_counts(normalize=True).unstack()
print("\nReadmission Rates by Age Group:\n", readmission_by_age)
```

```python
#analysing heatlcare utilisation across racial group

# 1. Time in Hospital by Race
plt.figure(figsize=(10, 6))
sns.boxplot(x='race', y='time_in_hospital', data=df)
plt.title('Time in Hospital by Race')
plt.xlabel('Race')
plt.ylabel('Time in Hospital')
plt.show()






```

```python
# 2. Number of Lab Procedures by Race
plt.figure(figsize=(10, 6))
sns.boxplot(x='race', y='num_lab_procedures', data=df)
plt.title('Number of Lab Procedures by Race')
plt.xlabel('Race')
plt.ylabel('Number of Lab Procedures')
plt.show()
```

```python
# 3. Number of Medications by Race
plt.figure(figsize=(10, 6))
sns.boxplot(x='race', y='num_medications', data=df)
plt.title('Number of Medications by Race')
plt.xlabel('Race')
plt.ylabel('Number of Medications')
plt.show()
```

```python
#The box plots help visualize differences in the distribution of time_in_hospital
#num_lab_procedures, and num_medications across different racial groups.
# 4. Outpatient, Emergency, and Inpatient Visits by Race
visit_cols = ['number_outpatient', 'number_emergency', 'number_inpatient']
df.groupby('race')[visit_cols].mean().plot(kind='bar', figsize=(10, 6))
plt.title('Healthcare Utilization by Race')
plt.xlabel('Race')
plt.ylabel('Mean Number of Visits')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Calculate and print descriptive statistics
print("\nDescriptive Statistics for Time in Hospital by Race:\n", df.groupby('race')['time_in_hospital'].describe())
print("\nDescriptive Statistics for Number of Lab Procedures by Race:\n", df.groupby('race')['num_lab_procedures'].describe())
print("\nMean Healthcare Utilization by Race:\n", df.groupby('race')[visit_cols].mean())
```
